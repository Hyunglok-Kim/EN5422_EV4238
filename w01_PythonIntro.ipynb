```python
# 1. Load the dataset from the nycflights13 package
flights = nycflights13.flights

# 1.1. Check the dataframe
flights
```

```python
# 2. Filter flights with a distance less than 1000 miles
short_flights = flights[flights['distance'] < 1000]

# 2.1. Check the dataframe
short_flights
```

```python
# 3. Keep only the specified columns
columns_of_interest = ['dep_delay', 'arr_delay', 'origin', 'dest', 'air_time', 'distance']
short_flights = short_flights[columns_of_interest]
# 3.1. Check the dataframe
short_flights
```

```python
# 4. Add the Z-score for departure delays
mean_delay = short_flights['dep_delay'].mean()
std_delay = short_flights['dep_delay'].std()
short_flights['dep_delay_z'] = (short_flights['dep_delay'] - mean_delay) / std_delay
# 4.1. Cehck the dataframe
short_flights
```

```python
# 5. Convert departure and arrival delays into hours
short_flights['dep_delay'] = short_flights['dep_delay'] / 60
short_flights['arr_delay'] = short_flights['arr_delay'] / 60
# 5.1. Cehck the dataframe
short_flights
```

```python
# 6. Calculate the average flight speed in mph
short_flights['avg_speed_mph'] = short_flights['distance'] / (short_flights['air_time'] / 60)
# 6.1. Check the dataframe
short_flights
```

```python
# 7. Order by average flight speed (fastest to slowest)
sorted_flights = short_flights.sort_values(by='avg_speed_mph', ascending=False)
# 7.1. Check the dataframe
sorted_flights
```

```python
# 8. Return the first 12 rows
top12_flights = sorted_flights.head(12)
print(top12_flights)
```

# 2.7 Groupwise Example

```python
import pandas as pd

# 1. Generate Sample dataset
data = {
    'location': ['Beach', 'Forest', 'Urban Park', 'Desert', 'Beach', 'Forest', 'Urban Park', 'Desert'],
    'parameter': ['Air Quality', 'Air Quality', 'Water Quality', 'Soil pH', 'Air Quality', 'Air Quality', 'Water Quality', 'Soil pH'],
    'date': ['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04', '2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04'],
    'value': [50, 45, 90, 7, 25, 20, 6.5, 85]
}
environmental_data = pd.DataFrame(data)

# 1.1. Check the dataframe
environmental_data
```

```python
# 2. Split: Group by 'location' and 'parameter'
grouped = environmental_data.groupby(['location', 'parameter'])
# 2.1. Check the dataframe
grouped
```

```python
# 3. Apply: Calculate average value for each group
average_values = grouped['value'].mean().reset_index()
# 3.1. Check the dataframe
average_values
```

```python
# 3.2. Rename columns for clarity
average_values.columns = ['location', 'parameter', 'average_value']
# 3.3. Check the dataframe
average_values
```

```python
# 4. Combine: In this case, the 'average_values' DataFrame is already a combined result
print(average_values)
```

#2.9 Iteration

```python
import numpy as np
import pandas as pd

# Simulate fake data
np.random.seed(2022)
data = pd.DataFrame({'x': np.random.rand(100), 'y': np.random.randn(100)})

# Fake model output
def model_1(x):
    return np.random.normal(loc=-0.5, scale=1, size=len(x))

def model_2(x):
    return np.random.normal(loc=0.5, scale=2, size=len(x))

# Using a for loop to simulate and assess performance in each subset
n_subsets = 5
np.random.seed(876)
output = []

for i in range(n_subsets):
    # Sub-sample data (25 samples)
    data_sub = data.sample(25)

    # Get output from the models
    yhat_1 = model_1(data_sub['x'])
    yhat_2 = model_2(data_sub['x'])

    # Score models (using MSE)
    perf_1 = np.mean((yhat_1 - data_sub['y'])**2)
    perf_2 = np.mean((yhat_2 - data_sub['y'])**2)

    # Save results
    output.append({'perf_1': perf_1, 'perf_2': perf_2, 'iter': i + 1})

# Convert to DataFrame and summarize
result_df = pd.DataFrame(output)
avg_diff = result_df['perf_1'].mean() - result_df['perf_2'].mean()
summary = pd.DataFrame({'avg.diff': [avg_diff], 'n': [len(result_df)]})

print(summary)
```

# 2.11 Iteration vs Vecotrization

```python
import numpy as np
import time

# Using Iteration
def iteration_example(arr):
    result = []
    for num in arr:
        result.append(num * 2)
    return result

# Using Vectorization
def vectorization_example(arr):
    return arr * 2

# Create a large array for demonstration
arr = np.random.rand(1000000)

# Measure time for iteration
start_time = time.time()
iteration_result = iteration_example(arr)
iteration_time = time.time() - start_time

# Measure time for vectorization
start_time = time.time()
vectorization_result = vectorization_example(arr)
vectorization_time = time.time() - start_time

print("Iteration time:", iteration_time)
print("Vectorization time:", vectorization_time)
```

#2.12 Iterate over elements of a list or array with Python's map()

```python
import numpy as np
import pandas as pd

def calculate_mse(data_sub, model_1, model_2):
    yhat_1 = model_1(data_sub['x'].values)
    yhat_2 = model_2(data_sub['x'].values)

    perf_1 = np.mean((yhat_1 - data_sub['y'].values) ** 2)
    perf_2 = np.mean((yhat_2 - data_sub['y'].values) ** 2)

    return pd.DataFrame({'perf_1': [perf_1], 'perf_2': [perf_2]})
```
