https://www.kaggle.com/jahnveenarang/cvdcvd-vd


'User ID': unique identification for consumer

'Age': cutomer age in years

'Estimated Salary': Avg. Income of consumer

'Gender': Whether consumer was male or female

'Purchased': 0 or 1 indicated clicking on Ad

```python
Ads_clicked_data = pd.read_csv('../input/cvdcvd-vd/Social_Network_Ads.csv')
Ads_clicked_data.head()
```

```python
Ads_clicked_data.drop(['User ID'],axis=1,inplace=True)
```

```python
Ads_clicked_data.head()
```

```python
Ads_clicked_data.isnull().sum()
```

```python
Ads_clicked_data.info()
```

```python
Ads_clicked_data.describe()
```

```python
fg,ax = plt.subplots(1,2,figsize=(18,6))
sns.distplot(Ads_clicked_data.Age,ax=ax[0]);
sns.distplot(Ads_clicked_data.EstimatedSalary,ax=ax[1]);
```

```python
sns.scatterplot(x='Age',y='EstimatedSalary',data=Ads_clicked_data,hue='Purchased',style='Purchased');
```

```python
sns.barplot(x='Gender',y='EstimatedSalary',data=Ads_clicked_data,hue='Purchased');
```

```python
sns.swarmplot(x='Gender',y='EstimatedSalary',data=Ads_clicked_data,hue='Purchased');
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler

LE = LabelEncoder()

Ads_clicked_data['Gender'] =  LE.fit_transform(Ads_clicked_data['Gender'])
```

```python
for col_name in Ads_clicked_data[['Age','EstimatedSalary']]:
  SS = StandardScaler()
  Ads_clicked_data[col_name] = SS.fit_transform(Ads_clicked_data[[col_name]])
```

```python
Ads_clicked_data = Ads_clicked_data.sample(frac=1.0)
```

```python
sns.heatmap(Ads_clicked_data.corr(),annot=True);
```

```python
from sklearn.model_selection import train_test_split

X = Ads_clicked_data[['Age','EstimatedSalary']]
Y = Ads_clicked_data['Purchased']

X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=10)

```

```python
from sklearn.tree import DecisionTreeClassifier

DTC = DecisionTreeClassifier()
DTC.fit(X_train,Y_train)

Y_pred = DTC.predict(X_test)
```

```python
from sklearn import metrics

print(metrics.accuracy_score(Y_test,Y_pred) * 100)
print(metrics.confusion_matrix(Y_test,Y_pred))
print(metrics.classification_report(Y_test,Y_pred))
```

# Decision Tree Regressor

```python
Mall_Customers_Data = pd.read_csv('../input/123qweasd/Mall_Customers.csv')
Mall_Customers_Data.head()
```

```python
Mall_Customers_Data.drop('CustomerID',axis=1,inplace=True)
Mall_Customers_Data.head()
```

```python
Mall_Customers_Data.isnull().sum()
```

```python
Mall_Customers_Data.info()
```

```python
fg,ax = plt.subplots(1,2,figsize=(18,6))
sns.distplot(Mall_Customers_Data.Age,ax=ax[0]);
sns.distplot(Mall_Customers_Data['Annual Income (k$)'],ax=ax[1]);
```

```python
sns.kdeplot(x='Age',y='Annual Income (k$)',data=Mall_Customers_Data);
```

```python
sns.scatterplot(x='Age',y='Spending Score (1-100)',hue='Annual Income (k$)',data=Mall_Customers_Data);
```

```python
from sklearn.preprocessing import LabelEncoder, RobustScaler

LE = LabelEncoder()

Mall_Customers_Data['Gender'] =  LE.fit_transform(Mall_Customers_Data['Gender'])
```

```python
for col_name in Mall_Customers_Data[['Age','Annual Income (k$)']]:
  RS = RobustScaler()
  Mall_Customers_Data[col_name] = RS.fit_transform(Mall_Customers_Data[[col_name]])
```

```python
sns.heatmap(Mall_Customers_Data.corr(),annot=True);
```

```python
from sklearn.model_selection import train_test_split

X = Mall_Customers_Data.iloc[:,:-1]
Y = Mall_Customers_Data['Spending Score (1-100)']

X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=10)
```

```python
from sklearn.tree import DecisionTreeRegressor

DTR = DecisionTreeRegressor()
DTR.fit(X_train,Y_train)

Y_pred = DTR.predict(X_test)
```

```python
from sklearn import metrics
import numpy as np

print(metrics.mean_absolute_error(Y_test,Y_pred))
print(metrics.mean_squared_error(Y_test,Y_pred))
print(np.sqrt(metrics.mean_squared_error(Y_test,Y_pred)))
```
