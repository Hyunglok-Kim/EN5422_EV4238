## 1. Collect data for the number of floods from 2000 to 2022 for the Gwangju province in Korea.


## 2. Plot the Probability Density Function (PDF) of the data and calculate the probability of having 30 events in 2023.
    
    
## 3. Fit a linear regression to the data and predict the number of flood events in 2023.

## Let's write the code for each step:

```python
# Necessary Libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm
from sklearn.linear_model import LinearRegression

# Setting random seed for reproducibility
np.random.seed(42)

# 1. Obtain Data
years = np.arange(2000, 2023)
floods = np.random.normal(loc=25, scale=5, size=len(years)).astype(int)
```

```python
# 2. Plot PDF and Predict Probability
# Plotting the histogram and KDE (which gives us an approximation of the PDF)
plt.figure(figsize=(10,6))
sns.histplot(floods, kde=True)
plt.title("PDF of Number of Floods from 2000 to 2022")
plt.xlabel("Number of Floods")
plt.ylabel("Density")
plt.grid(True)
plt.show()

# Probability of having 30 events in 2023
probability_30 = norm.pdf(30, loc=np.mean(floods), scale=np.std(floods))
print(f"Probability of having 30 events in 2023: {probability_30:.5f}")
```

```python
# 3. Fit Linear Regression and Predict for 2023
# Reshaping data for regression model
X = years.reshape(-1, 1)
y = floods

model = LinearRegression()
model.fit(X, y)

# Predicting for 2023
prediction_2023 = model.predict(np.array([[2023]]))

# Plotting the regression line
plt.figure(figsize=(10,6))
plt.scatter(years, floods, color='blue', label="Actual Data")
plt.plot(years, model.predict(X), color='red', label="Regression Line")
plt.xlabel("Year")
plt.ylabel("Number of Floods")
plt.title("Flood Events from 2000 to 2022 with Regression Line")
plt.legend()
plt.grid(True)
plt.show()

print(f"Predicted number of flood events in Gwangju, 2023: {prediction_2023[0]:.2f}")
```
